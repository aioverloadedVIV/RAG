```python
# Let's make it a function
def word_tokenization(text: str):
    # convert the text into lower letters
    text_lower = text.lower()
    
    # implement the word tokenizer
    tokens = nltk.word_tokenize(text=text_lower)

    # remove punctuations and stopwords
    final_tokens = [word for word in tokens if word.isalnum()]
    return final_tokens
```

### 1. Vector Space Model (TF-IDF)

tf (t,d) = No. of times term 't' appears in document 'd' / Total number of terms in document d

IDF (t) = log (Total number of documents / Number of documents containing term 't')

Tf-IDF = TF * IDF

```python
# Creating an instance of the TF-IDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the preprocessed docs
tf_idf_matrix = vectorizer.fit_transform(preprocessed_docs)
print(f"The shape of the TF-IDF matrix is: {tf_idf_matrix.shape}")
print(f"The length of the TF-IDF matrix is: {len(documents)}")

# Query the index
query = "croatia sailing"
query_vec = vectorizer.transform([query])
similarities = cosine_similarity(tf_idf_matrix, query_vec).flatten()
print(f"\nShape of the query vector: {query_vec.shape}\n")
print(similarities)
```

```python
# Sort the documents by similarity to the queries and checking which part is more similar
similarities = cosine_similarity(tf_idf_matrix, query_vec).flatten()
sorted_similarities = list(enumerate(similarities))

sorted(sorted_similarities, key=lambda x: x[1], reverse=True)
```

```python
# --- Build a function to search with Tf-IDF --- 
def search_tfidf(query, vectorizer, tfidf_matrix):
    """ It searches the query inside the tfidf_matrix to find the most similar results using cosine similarity.
    
    Args:
        query: user query 
        vectorizer : TfidfVectorizer()
        tfidf_matrix: The shape of the TF-IDF matrix is: (100, 360)
    
    Returns:
        results : most similar 

    """

    # vectorizing the query
    query_vec = vectorizer.transform([query])

    # Compute the cosine similarity
    similarities = cosine_similarity(tfidf_matrix, query_vec).flatten()

    # Pair each document index with it's similarity score
    sorted_similarities = list(enumerate(similarities))

    # Sort the documents by similarity score
    results = sorted(sorted_similarities, key=lambda x: x[1], reverse=True)
    return results 
```

```python
# Applying the function to the query
search_similarities = search_tfidf(query, vectorizer, tf_idf_matrix)
#print(search_similarities)

# --- Print all the matching documents with query by similarity score ---
#print(f"Fetch all the documents by similarity score for query {query}: ")
#for doc_index, score in search_similarities:
    #print(f"Document {doc_index+1}: {documents[doc_index]}")

# --- Print only top 10 matching documents with query by similarity score ---
print(f"Fetch all the documents by similarity score for query {query}: ")
for doc_index, score in search_similarities[:10]:
    print(f"Document {doc_index+1}: {documents[doc_index]}")
```

OR: Retrieves documents containing atleast one of given terms. 

NOT: Omits documents with specified term.
